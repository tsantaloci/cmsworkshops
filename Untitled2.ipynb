##Plotting Lesson

```python
import numpy
import os
```

```python
distance_file = os.path.join('data','distance_data_headers.csv')
distances = numpy.genfromtxt(fname=distance_file, delimiter=',',dtype='unicode')
print(distances)
```

```python
headers = distances[0] ##slicing
print(headers)
```

```python
data= distances[1:]
print(data)
data=data.astype(numpy.float)
```

```python
import matplotlib.pyplot as plt ##library name as nickname for library
```

```python
plt.figure()
plt.plot(data[:,0],data[:,1]) # plotting every row and index 1 column
```

```python
sample = headers[1]

plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
fig_1 = plt.plot(data[:,0], data[:,1], label=sample)
plt.legend()

plt.savefig(F'{sample}.png')
```

```python
plt.figure()
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstroms)')
plt.plot(data[:,0],data[:,1], label=headers[1])
plt.plot(data[:,0],data[:,2], label=headers[2])
plt.legend()
```
